Fashion_MSNIST dataset

```python
from keras.datasets import fashion_mnist
```

This is a dataset of 60,000 28x28 grayscale images of 10 fashion categories, along with a test set of 10,000 images. This dataset can be used as a drop-in replacement for MNIST.

The classes are:

Label	Description

0	T-shirt/top

1	Trouser

2	Pullover

3	Dress

4	Coat

5	Sandal

6	Shirt

7	Sneaker

8	Bag

9	Ankle boot

```python
label_map = {0: 'T-shirt/top', 1: 'Trouser', 2: 'Pullover', 3: 'Dress', 
             4: 'Coat', 5: 'Sandal', 6: 'Shirt', 7: 'Sneaker', 8: 'Bag', 9: 'Ankle boot'}
```

```python
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

The values are between 0-255. Normalize the value between 0-1

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train[0]
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Flatten
from keras import metrics
```

```python
model = Sequential()
```

```python
# Add inout layer
# Flat the 28x28 image to a 1-D array using Flatten
model.add(Flatten(input_shape=[28, 28]))

# Hidden layer 20 neurons with activation function relu
model.add(Dense(20, activation='relu'))

# add output layer 10 classes
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss = 'sparse_categorical_crossentropy',
             optimizer = 'adam',
             metrics = ['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=10)
```

```python
model.evaluate(X_test, y_test)
```

```python
plt.imshow(X_test[0])
```

```python
y_predict = model.predict(X_test)
```

```python
y_predict[0]

# This will give the probablities in all the classes
```

```python
# Find the max problatity class
max_prob_class = np.argmax(y_predict[0])
max_prob_class
```

```python
label_map[max_prob_class]
```
